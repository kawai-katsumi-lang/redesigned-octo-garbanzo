```python
# 1. 必要なライブラリ（Whisper）をインストール
# ※実行ボタンを押すと、完了まで少し時間がかかります
!pip install git+https://github.com/openai/whisper.git
!sudo apt update && sudo apt install ffmpeg

import whisper
from google.colab import files

# 2. 高精度モデル(large)をロード
# ※GPUモードでないとここで時間がかかります
print("AIモデルを準備中...")
model = whisper.load_model("large")

# 3. 音声ファイルをアップロード
print("\n--- 音声ファイルのアップロード ---")
print("下の「ファイルを選択」ボタンを押して、文字起こししたいmp3ファイルを選んでください。")
uploaded = files.upload()

# 4. 文字起こし実行
for filename in uploaded.keys():
    print(f"\nProcessing: {filename} ...")

    # 日本語として認識・テキスト化
    result = model.transcribe(filename, language="ja")

    print("\n" + "="*30)
    print(f"▼ {filename} の文字起こし結果 ▼")
    print("="*30)
    print(result["text"])
    print("="*30 + "\n")

    # (オプション) テキストファイルとして保存・ダウンロードする場合
    txt_filename = filename + ".txt"
    with open(txt_filename, "w", encoding="utf-8") as f:
        f.write(result["text"])
    files.download(txt_filename)
```
