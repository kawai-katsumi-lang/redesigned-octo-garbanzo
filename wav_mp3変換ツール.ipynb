```python
# =================================================================
# Google Colabå°‚ç”¨: WAV to MP3 ä¸€æ‹¬å¤‰æ›ã‚¹ã‚¯ãƒªãƒ—ãƒˆ
# ã“ã®ã‚³ãƒ¼ãƒ‰å…¨ä½“ã‚’ã‚³ãƒ”ãƒ¼ã—ã¦ã€Colabã®ã‚»ãƒ«ã«è²¼ã‚Šä»˜ã‘ã¦å®Ÿè¡Œã—ã¦ãã ã•ã„(â–¶ãƒœã‚¿ãƒ³)ã€‚
# =================================================================

import os
import subprocess
import sys
import shutil
import time

def install_dependencies():
    """å¿…è¦ãªãƒ©ã‚¤ãƒ–ãƒ©ãƒªã¨ãƒ„ãƒ¼ãƒ«(ffmpeg, pydub)ã‚’ã‚¤ãƒ³ã‚¹ãƒˆãƒ¼ãƒ«ã—ã¾ã™"""
    print("âš™ï¸ ç’°å¢ƒæ§‹ç¯‰ä¸­... (åˆå›ã¯æ™‚é–“ãŒã‹ã‹ã‚‹å ´åˆãŒã‚ã‚Šã¾ã™)")

    # FFmpegã®ã‚¤ãƒ³ã‚¹ãƒˆãƒ¼ãƒ« (apt-getã‚’ä½¿ç”¨)
    if not shutil.which("ffmpeg"):
        print("   - FFmpegã‚’ã‚¤ãƒ³ã‚¹ãƒˆãƒ¼ãƒ«ã—ã¦ã„ã¾ã™...")
        subprocess.run(["apt-get", "install", "-y", "ffmpeg"], check=True)

    # pydubã®ã‚¤ãƒ³ã‚¹ãƒˆãƒ¼ãƒ« (pipã‚’ä½¿ç”¨)
    try:
        import pydub
    except ImportError:
        print("   - pydubãƒ©ã‚¤ãƒ–ãƒ©ãƒªã‚’ã‚¤ãƒ³ã‚¹ãƒˆãƒ¼ãƒ«ã—ã¦ã„ã¾ã™...")
        subprocess.check_call([sys.executable, "-m", "pip", "install", "pydub"])

def main():
    install_dependencies()

    # Google Colabå›ºæœ‰ã®ãƒ¢ã‚¸ãƒ¥ãƒ¼ãƒ«ã‚’ã‚¤ãƒ³ãƒãƒ¼ãƒˆ
    try:
        from google.colab import files
        from pydub import AudioSegment
    except ImportError:
        print("âŒ ã‚¨ãƒ©ãƒ¼: Google Colabç’°å¢ƒã§ã¯ã‚ã‚Šã¾ã›ã‚“ã€‚")
        return

    print("\nğŸ“¤ å¤‰æ›ã—ãŸã„ .wav ãƒ•ã‚¡ã‚¤ãƒ«ã‚’ã‚¢ãƒƒãƒ—ãƒ­ãƒ¼ãƒ‰ã—ã¦ãã ã•ã„ (è¤‡æ•°é¸æŠå¯)")
    uploaded = files.upload()

    if not uploaded:
        print("âš ï¸ ãƒ•ã‚¡ã‚¤ãƒ«ãŒã‚¢ãƒƒãƒ—ãƒ­ãƒ¼ãƒ‰ã•ã‚Œã¾ã›ã‚“ã§ã—ãŸã€‚")
        return

    # ä¿å­˜ç”¨ãƒ•ã‚©ãƒ«ãƒ€ã®æº–å‚™
    output_dir = "mp3_output"
    if os.path.exists(output_dir):
        shutil.rmtree(output_dir)
    os.makedirs(output_dir)

    print("\nğŸš€ å¤‰æ›ã‚’é–‹å§‹ã—ã¾ã™...")
    converted_count = 0

    for filename in uploaded.keys():
        if not filename.lower().endswith(".wav"):
            print(f"â© ã‚¹ã‚­ãƒƒãƒ— (wavã§ã¯ã‚ã‚Šã¾ã›ã‚“): {filename}")
            continue

        try:
            # èª­ã¿è¾¼ã¿
            audio = AudioSegment.from_wav(filename)

            # å‡ºåŠ›ãƒ•ã‚¡ã‚¤ãƒ«å
            mp3_filename = os.path.splitext(filename)[0] + ".mp3"
            output_path = os.path.join(output_dir, mp3_filename)

            # æ›¸ãå‡ºã—
            audio.export(output_path, format="mp3", bitrate="192k")
            print(f"âœ… å¤‰æ›å®Œäº†: {mp3_filename}")
            converted_count += 1

        except Exception as e:
            print(f"âŒ å¤‰æ›ã‚¨ãƒ©ãƒ¼ ({filename}): {e}")

    if converted_count > 0:
        print(f"\nğŸ“¦ MP3ãƒ•ã‚¡ã‚¤ãƒ«ã‚’ZIPã«åœ§ç¸®ã—ã¦ã„ã¾ã™...")
        zip_filename = "converted_mp3_files.zip"
        shutil.make_archive("converted_mp3_files", 'zip', output_dir)

        print(f"ğŸ“¥ ãƒ€ã‚¦ãƒ³ãƒ­ãƒ¼ãƒ‰ã‚’é–‹å§‹ã—ã¾ã™: {zip_filename}")
        files.download(zip_filename)
        print("ğŸ‰ ã™ã¹ã¦å®Œäº†ã—ã¾ã—ãŸï¼")
    else:
        print("âš ï¸ å¤‰æ›ã•ã‚ŒãŸãƒ•ã‚¡ã‚¤ãƒ«ã¯ã‚ã‚Šã¾ã›ã‚“ã§ã—ãŸã€‚")

if __name__ == "__main__":
    main()
```
